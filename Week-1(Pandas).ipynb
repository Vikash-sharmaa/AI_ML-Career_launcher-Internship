### `Query 1.4`
In a separate array , calculate the monthwise VWAP (Volume Weighted Average Price ) of the stock.
( VWAP = sum(price*volume)/sum(volume) )
To know more about VWAP , visit - VWAP definition
{Hint : Create a new dataframe column ‘Month’. The values for this column can be derived from the ‘Date” column by using appropriate pandas functions. Similarly, create a column ‘Year’ and initialize it. Then use the 'groupby()' function by month and year. Finally, calculate the vwap value for each month (i.e. for each group created).


```python
df['Month']=pd.DatetimeIndex(df['Date']).month
df['Year']=pd.DatetimeIndex(df['Date']).year
df['VWAP'] = (df['Close Price'] * df['Total Traded Quantity']).cumsum()/df['Total Traded Quantity'].fillna(0).cumsum()
df_vwap = df[['Month','Year','VWAP']]
group = df_vwap.groupby(['Month','Year'])
group.first().head(10)
```

### `Query 1.5`
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def average(n):
    df1=df["Close Price"].iloc[-1:-(n+1):-1].mean()
    return df1
def profit_loss(N):
    change = (df['Close Price'].tail(N).iloc[N-1] - df['Close Price'].tail(N).iloc[0])
    if change < 0 :
        loss = -(change)
        loss_percen = (loss/df['Close Price'].tail(N).iloc[N-1])*100
        return loss_percen
    if change > 0 :
        profit = change
        profit_percen = (profit/df['Close Price'].tail(N).iloc[N-1])*100
        return profit_percen
df2=pd.DataFrame({'Average':[average(5),average(10),average(20),average(60),average(120),average(240)],'Profit Loss':[profit_loss(5),profit_loss(10),profit_loss(20),profit_loss(60),profit_loss(120),profit_loss(240)]},index=['Last 1 Week','Last 2 Weeks','Last 1 Months','Last 3 Months','Last 6 Months','Last 1 Year'])
df2.head()
```

### `Query 1.6`
Add a column 'Day_Perc_Change' where the values are the daily change in percentages i.e. the percentage change between 2 consecutive day's closing prices. Instead of using the basic mathematical formula for computing the same, use 'pct_change()' function provided by Pandas for dataframes. ### `Query 1.7`
Add another column 'Trend' whose values are: 'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5 'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1 'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1 'Positive' for 'Day_Perc_Change' in between 1 and 3 'Negative' for 'Day_Perc_Change' in between -1 and -3 'Among top gainers' for 'Day_Perc_Change' in between 3 and 7 'Among top losers' for 'Day_Perc_Change' in between -3 and -7 'Bull run' for 'Day_Perc_Change' >7 'Bear drop' for 'Day_Perc_Change' <-7

```python
df.loc[(df['Day_Perc_Change'] >= -0.5) & (df['Day_Perc_Change'] <= 0.5), 'Trend'] ="Slight or No change"
df.loc[(df['Day_Perc_Change'] >= 0.5) & (df['Day_Perc_Change'] <= 1), 'Trend'] ="Slight Positive"
df.loc[(df['Day_Perc_Change'] <= -0.5) & (df['Day_Perc_Change'] >= -1), 'Trend'] ="Slight Negative"
df.loc[(df['Day_Perc_Change'] >= 1) & (df['Day_Perc_Change'] <= 3), 'Trend'] ="Positive"

df.loc[(df['Day_Perc_Change'] <= -1) & (df['Day_Perc_Change'] >= -3), 'Trend'] ="Negative"
df.loc[(df['Day_Perc_Change'] >= 3) & (df['Day_Perc_Change'] <= 7), 'Trend'] ="Among top gainers"
df.loc[(df['Day_Perc_Change'] <= -3) & (df['Day_Perc_Change'] >= -7), 'Trend'] ="Among top losers"
df.loc[(df['Day_Perc_Change'] > 7), 'Trend'] ="Bull run"
df.loc[(df['Day_Perc_Change'] < -7), 'Trend'] ="Bear drop"
df
```
